## Introduction

```python
import pandas as pd
```

```python
#data - python variable - referes to the dict
#labels of columns - name. city, age, py score
# row_labels - labels teh rows

data = {
    'name': ["Xavier", "Ann", "Jana", "Yi", "Robin", "Amal", "Nori"],
    'city': ["Mexico", "Toronto", "Prague", "Shanghai", "Manchester", "Hyderabad", "Chennai"],
    'age' : [41,28,33,34,38,31,37],
    'py-score': [88.0, 79.0, 81.0, 80.0, 68.0, 61.0, 84.0]
}

row_labels = [101,102,103,104,105,106,107]
```

```python
#create a pandas dataframe:

df = pd.DataFrame(data= data, index = row_labels)
df
```

```python
# Pandas Datafreames can be very large, making it impractical to look at once.
# we can use  .head() to show first few items

df.head(n=2)   # parameter n specifies the number of rows to show
```

```python
# we can use .tail() to show last few items

df.tail(n=2)   # parameter n specifies the number of rows to show
```

```python
# we can access a column in a pandas Dataframe the same way you would get a value from dict:
cities = df["city"]
cities
```

```python
#if the name of the column is a string , then we can use a dot notation to access it

df.city
```

```python
# Each column of a pandas DataFrame is an instance of pandas.Series, a structure that holds one-dimensional data and their labels.
# You can get a single item of a Series object the same way you would with a dictionary, by using its label as a key:

cities[102]
```

```python
# we can also access a whole row with accessor - .loc[]
df.loc[103]
```

### Creating a pandas DataFrame with Dictionaries


```python
# We cna create a pandas DataFrame with Python Dict:

d = {'x' : [1,2,3], 'y' : np.array([2,4,8]), 'z': 100}

pd.DataFrame(d)
```

```python
# it is possible to control the order of the columns with columns paramater and the row labels with index:

pd.DataFrame(d, index = [100,200,300], columns = ['z', 'y', 'x'])
```

###Creating a pandas DataFrame with Lists

```python
a = [{'x': 1, 'y': 2, 'z': 100},
     {'x': 2, 'y': 4, 'z': 100},
     {'x': 3, 'y': 8, 'z': 100}]


pd.DataFrame(a)
```

```python
# Nested List

a = [[1,2,100],
     [2,4,100],
     [3,8,100]]

pd.DataFrame(a, columns = ['x','y','z'])
```

###Creating a pandas DataFrame with NumPy Arrays

```python
arr = np.array([[1,2,100],
                [2,4,100],
                [3,8,100]])

df_ = pd.DataFrame(arr, columns = ['x', 'y', 'z'])

df_
```

```python
# You can specify the optional parameter copy.

arr[0,0] = 1000

df_
```
